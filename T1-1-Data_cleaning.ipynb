### Parametros read_csv
- filepath: ruta completa
- sep: separador
- dtype: tipo de los datos por columnas ({"a":np.int32...})
- header: la cabecera de los datos(none=0 -> primera fila)
- names: nombres de las columnas
- skiprows: saltarse alguna fila
- index_col: elegir la columna como "clasificador" del datset
- skip_blank_lines: evitar espacios en blanco(en vez de NaN poner en blanco)
- na_filter: eliminar los valores NaN(toda la fila)

##### Ejemplo:

# Carga de datos a través de la funcion open

## Lectura de ficheros

```python
data3 = open(mainpath + "/" + "customer-churn-model/Customer Churn Model.txt", "r")
```

```python
cols = data3.readline().strip().split(",")
n_cols = len(cols)
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
for line in data3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter+=1
    
print ("El dataset tiene %d filas y %d columnas"%(counter, n_cols))
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

## Lectura y escritura de ficheros

```python
infile = mainpath + "/" + "customer-churn-model/Customer Churn Model.txt"
outfile = mainpath + "/" + "customer-churn-model/Tab Customer Churn Model.txt"
```

```python
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```

```python
df4 = pd.read_csv(outfile,sep="\t")
df4.head()
```
