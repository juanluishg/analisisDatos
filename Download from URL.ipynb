```python
def downloadFromUrl(url, filename, mainpath = "/home/juanluishg/Documentos/python-ml-course/datasets",
                    sep = ",", delim = "\n", encoding = "utf-8"):
    
    #Importar librerias necesarias
    import pandas as pd
    import urllib3
    import csv
    import os
    
    #Nos conectamos a la url y cogemos los datos
    http = urllib3.PoolManager()
    r = http.request("GET", url)
    r.status
    response = r.data

    #Leemos los datos de la respuesta
    cr = csv.reader(response)
    
    #Los datos estan en String Binarios(codificados), los descodificamos y separamos
    str_decode = response.decode(encoding)
    data = str_decode.strip().split(delim)
    
    #Nos quedamos con la cabezera
    cols = data[0].split(sep)
    n_cols = len(cols)
     
    main_dict = {}
    counter = 0
    
    for col in cols:
        main_dict[col]=[]
    
    for line in data:
        if(counter>0):
            values = line.strip().split(sep)
            for i in range(len(cols)):
                main_dict[cols[i]].append(values[i])
        counter+=1
        
    df = pd.DataFrame(main_dict)
    
    fullpath = os.path.join(mainpath, filename)
    
    df.to_csv(fullpath + ".csv")
    df.to_json(fullpath + ".json")
    
    print("Los ficheros se han guardado correctamente en: " + fullpath)
    
    return df
```
